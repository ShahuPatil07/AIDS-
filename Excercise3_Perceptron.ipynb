```python
import numpy as np
import matplotlib.pyplot as plt


x = np.random.randint(1, 100, size=(50))
y = np.ones(50)
y[x <= 50] = -1


for i in range(50):
    color = 'blue' if y[i] == 1 else 'red'
    plt.scatter(x[i], 0, color=color)

wo = 0
w1 = 1

misc = 100
learning_rate = 0.1

while misc != 0:
    misc = 0

    for i in range(50):
        if (w1 * x[i] + wo) * y[i] <= 0:
            misc += 1
            wo = wo + y[i] * learning_rate
            w1 = w1 + y[i] * x[i] * learning_rate

x_decision = np.linspace(0, 100, 100)
y_decision = (-w1 * x_decision - wo) / w1
plt.plot(x_decision, y_decision, color='green', linestyle='--', label='Decision Boundary')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.random.randint(0, 100, size=(50))
x2 = np.random.randint(0, 100, size=(50))

y = []
for i in range(50):
    if x1[i] > x2[i]:
        y.append(1)
    else:
        y.append(-1)

for i in range(50):
    color = 'blue' if x1[i] > x2[i] else 'red'
    plt.scatter(x1[i], x2[i], color=color)



wo = 0
w1 = 1
w2 = 9

misc = 100
learning_rate = 0.1

while misc != 0:
    misc = 0

    for i in range(50):
        if (w1 * x1[i] + w2 * x2[i] + wo) * y[i] <= 0:
            misc += 1
            wo = wo + y[i] * learning_rate
            w1 = w1 + y[i] * x1[i] * learning_rate
            w2 = w2 + y[i] * x2[i] * learning_rate

print("Final weights:", w1, w2, wo)
x_decision = np.linspace(0, 100, 100)
y_decision = (-w1 * x_decision - wo) / w2
plt.plot(x_decision, y_decision, color='green', linestyle='--', label='Decision Boundary')
plt.show()
```
